```python
%%writefile app.py

import streamlit as st
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_google_genai import ChatGoogleGenerativeAI
import os
import google.generativeai as genai

# Set up Google API key (ensure this is set in your Colab environment or directly here)
os.environ["GOOGLE_API_KEY"] = "AIzaSyBiBCiL8IcePe5HvG1IuIJgBGcseVEqldg" # Replace with your actual key if not set in Colab secrets
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

# Load and process the PDF document (re-run logic from previous cells)
def load_and_process_pdf():
    try:
        loader = PyPDFLoader("/content/Logisticsfin.pdf")
        documents = loader.load()

        text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=1000,
            chunk_overlap=200
        )
        chunks = text_splitter.split_documents(documents)

        embeddings = HuggingFaceEmbeddings(
            model_name="sentence-transformers/all-MiniLM-L6-v2"
        )
        vectorstore = FAISS.from_documents(chunks, embeddings)
        return vectorstore
    except Exception as e:
        st.error(f"Error loading or processing PDF: {e}")
        return None

# Initialize LLM
llm = ChatGoogleGenerativeAI(
    model="models/gemini-2.5-flash",
    temperature=0.3
)

# Define the ask_question function
def ask_question(query, vectorstore):
    if vectorstore is None:
        return "Document processing failed. Please ensure PDF is uploaded and try again."

    # 1. Retrieve relevant chunks
    docs = vectorstore.similarity_search(query, k=2)
    context = "\n\n".join([doc.page_content for doc in docs])

    # 2. Create prompt
    prompt = f"""
You are an AI assistant.
Answer the question strictly using the context below.
If the answer is not in the context, say "Answer not found in the document."

Context:
{context}

Question:
{query}

Answer:
"""

    # 3. Ask Gemini
    response = llm.invoke(prompt)
    return response.content

# Streamlit app layout
st.title("Logistics Chatbot")
st.write("Ask me anything")

# Load vectorstore once
@st.cache_resource
def get_vectorstore():
    return load_and_process_pdf()

vectorstore = get_vectorstore()

if vectorstore:
    query = st.text_input("Enter your question:")

    if st.button("Get Answer"):
        if query:
            with st.spinner("Finding the answer..."):
                answer = ask_question(query, vectorstore)
                st.write("**Answer:**")
                st.write(answer)
        else:
            st.warning("Please enter a question.")
else:
    st.error("Cannot proceed without a processed this document.")
```

## Define Ground Truth Data

### Subtask:
Create a set of ground truth questions and their corresponding correct answers to evaluate the model's accuracy.


**Reasoning**:
Define a list of dictionaries named `ground_truth_data` containing relevant questions and their corresponding answers from the `Logisticsfin.pdf` document to serve as ground truth for evaluation.



```python
ground_truth_data = [
    {'question': 'What is the primary function of logistics?', 'answer': 'The primary function of logistics is to manage the flow of goods, services, and information from the point of origin to the point of consumption.'},
    {'question': 'What are the key components of a supply chain?', 'answer': 'The key components of a supply chain typically include suppliers, manufacturers, warehouses, distribution centers, retailers, and customers.'},
    {'question': 'What role does transportation play in a supply chain?', 'answer': 'Transportation is a critical component of the supply chain, involving the movement of goods between various points such as suppliers, manufacturers, warehouses, and customers.'},
    {'question': 'What are the main objectives of supply chain management?', 'answer': 'The main objectives of supply chain management are to optimize the flow of goods and information, reduce costs, improve efficiency, enhance customer satisfaction, and achieve competitive advantage.'},
    {'question': 'How does inventory management impact logistics?', 'answer': 'Inventory management is crucial in logistics as it deals with the storage and control of goods, balancing the need to meet customer demand with the costs associated with holding inventory.'}
]

print(f"Ground truth data created with {len(ground_truth_data)} question-answer pairs.")
```

## Measure Response Time and Evaluate Answers

### Subtask:
For each question in the ground truth data, measure the time taken by the `ask_question` function to generate a response. Compare the model's answer with the ground truth answer.


**Reasoning**:
To measure the response time and collect evaluation data, I will iterate through the ground truth questions, record the time before and after calling the `ask_question` function, and store the results in a list.



```python
import time

evaluation_results = []

for item in ground_truth_data:
    question = item['question']
    ground_truth_answer = item['answer']

    start_time = time.time()
    model_answer = ask_question(question, vectorstore)
    end_time = time.time()
    response_time = end_time - start_time

    evaluation_results.append({
        'question': question,
        'ground_truth_answer': ground_truth_answer,
        'model_answer': model_answer,
        'response_time': response_time
    })

print("Evaluation Results:")
for result in evaluation_results:
    print(result)
```

**Reasoning**:
Now that the evaluation results are collected, I will calculate the average response time from the `evaluation_results` list and present it as part of the overall performance metrics.



```python
total_response_time = sum([res['response_time'] for res in evaluation_results])
average_response_time = total_response_time / len(evaluation_results)

print(f"\nAverage Response Time: {average_response_time:.2f} seconds")
```

**Reasoning**:
To evaluate the correctness of the model's responses, I will iterate through the `evaluation_results` and compare the `model_answer` with the `ground_truth_answer` for each question. I will then perform a basic keyword overlap check to assign a simple correctness score for each answer.



```python
import re

def calculate_correctness(model_answer, ground_truth_answer):
    # Convert both answers to lowercase for case-insensitive comparison
    model_answer_lower = model_answer.lower()
    ground_truth_answer_lower = ground_truth_answer.lower()

    # A very simple correctness check: if the ground truth answer is a substring of the model answer
    # or if there's significant keyword overlap.
    # For this simplified example, let's check if the model answer contains a significant portion of the ground truth answer's keywords.
    # This is a very basic heuristic and can be improved with more sophisticated NLP techniques.

    # Split ground truth into keywords, filter out common stop words and single characters
    gt_keywords = set(re.findall(r'\b\w+\b', ground_truth_answer_lower))
    gt_keywords = {word for word in gt_keywords if len(word) > 2 and word not in ['the', 'and', 'is', 'of', 'in', 'to', 'a', 'an', 'for', 'with']}

    # Split model answer into keywords
    model_keywords = set(re.findall(r'\b\w+\b', model_answer_lower))
    model_keywords = {word for word in model_keywords if len(word) > 2 and word not in ['the', 'and', 'is', 'of', 'in', 'to', 'a', 'an', 'for', 'with']}

    # Calculate overlap
    common_keywords = gt_keywords.intersection(model_keywords)

    # Consider an answer 'correct' if more than 50% of ground truth keywords are present in the model answer
    # and if the model answer is not 'Answer not found in the document.'
    if "answer not found in the document." in model_answer_lower:
        return 0 # Definitely incorrect if it couldn't find the answer

    if len(gt_keywords) == 0: # Avoid division by zero if ground truth has no significant keywords
        return 1 if len(model_keywords) > 0 else 0

    overlap_ratio = len(common_keywords) / len(gt_keywords)
    return 1 if overlap_ratio >= 0.5 else 0 # Simple binary correctness


correct_answers_count = 0
print("\n--- Correctness Evaluation ---")
for i, result in enumerate(evaluation_results):
    is_correct = calculate_correctness(result['model_answer'], result['ground_truth_answer'])
    result['is_correct'] = is_correct
    if is_correct:
        correct_answers_count += 1

    print(f"Question {i+1}: {result['question']}")
    print(f"  Ground Truth: {result['ground_truth_answer']}")
    print(f"  Model Answer: {result['model_answer']}")
    print(f"  Correctness: {'Correct' if is_correct else 'Incorrect'}")
    print(f"  Response Time: {result['response_time']:.2f} seconds\n")

total_questions = len(evaluation_results)
accuracy = (correct_answers_count / total_questions) * 100

print(f"Total Questions: {total_questions}")
print(f"Correct Answers: {correct_answers_count}")
print(f"Accuracy (based on keyword overlap): {accuracy:.2f}%")
```
